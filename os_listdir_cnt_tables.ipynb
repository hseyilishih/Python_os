```python
'''
計算某個資料夾轄下 每個子資料夾內的檔案個數與大小(mb)
子資料夾本身不算一個
網路磁碟機可用代號的命名 eg. Y:
'''
import os

#*******************************
os.chdir('d:\R')  #大小寫要相符     
#*******************************
rootdir = os.getcwd()

#*******************************
x  = u'd:\cnt_tables_20180325.csv'
#*******************************

f = open(os.path.join(rootdir,x), "w+")
f.write("folder_path" +'|'+ "file_count_in_folder" +'|' +  "total_size" +'|' + "size_unit"  +'|' 'folder_type' + '\n')

for dirname in  os.listdir(rootdir):   
    Dir = os.path.join(rootdir, dirname)    
    count_size = 0
    count_file = 0
    
    if (os.path.isdir(Dir)):           #如果類型是一個目錄
        print("\n", "Sub_folder=", Dir,"﹀"*10, "\n")
        for root, ds, files in os.walk(Dir): #目錄下所有目錄
            for file in files:     
                  
                count_file += 1
                # print(" "*5, count_file,"\t", os.path.splitext(file)[0],"\t", os.path.splitext(file)[-1])
                print(" "*5, count_file, "\t", file)
                size = os.path.getsize(os.path.join(root, file)) 
                count_size += size
                

        #if ((count/1024.0/1024.0) < 1024):
            #print ("*****", Dir +'\t' + '%.3f'% (count_size/1024.000/1024.000)+' MB'+ " cnt=" + str(count_file) + '\n')
        print("\n", "Sub_folder=", Dir , '-----' , '%.3f'% (count_size/1024.0/1024.0), "MB" ,"︿"*10, "\n")
        f.write(Dir +'|' + str(count_file)+ '|' +  '%.3f'% (count_size/1024.0/1024.0)+'|MB' + '|sub_folder'+ '\n')
        #else:
            #print ( "*****", Dir + '\t' + '%.3f' % (count_size / 1024.0 / 1024.0/1024.0) + ' GB' + " cnt=" + str(count_file) + '\n')
 
            #f.write(Dir +'|' + str(count_file)+ '|' + '%.3f' % (count_size / 1024.0 / 1024.0/1024.0) + '|GB' +'\n')

    else:
        print ("\n", "直轄=" + Dir + '-----' + '\t' + '%.3f' % (os.path.getsize(Dir)/1024.0/1024.0 ) +' MB' )
        f.write(Dir +'|' + '1' + '|' +  '%.3f'% (os.path.getsize(Dir)/1024.0/1024.0)+'|MB' + '|direct_file'+ '\n')
        continue
f.close()
```
